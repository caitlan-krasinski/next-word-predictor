```python
import re
import pandas as pd
data = pd.read_csv("IMDB Dataset.csv")
```

```python
data.head()
```

```python
data['review'][2]
```

```python
def clean_words(text):
    s = re.sub(re.compile('<.*?>'), ' ', text) #remove break tags
    s = re.sub(' +', ' ', s)
    s = re.sub("#","",s) #remove #
    s = re.sub(r'[0-9]+', '', s) #remove numbers
    s = re.sub("[^\w\d'\s]+",'',s) #remove punctuation but keep '
    s = s.lower()                        
    #words = lower_case.split()                              
    #return (" ".join(words)).strip()
    return s
```

```python
clean_words(data['review'][2])
```

```python
clean_reviews = []

for i in data.review:
    clean_reviews.append(clean_words(i))
```

```python
import nltk 
nltk.download('words')
w = set(nltk.corpus.words.words())
w
```

```python
clean_reviews = clean_reviews[:10000]
len(clean_reviews)
```

```python
#remove non english words / typos

clean_reviews2 = []
for review in clean_reviews:
    words = review.split()
    for word in words:
        if word not in w:
            words.remove(word)

    x = " ".join(words)
    clean_reviews2.append(x)
        
    
len(clean_reviews2)
```

```python
clean_data = pd.DataFrame(data=clean_reviews2, index=None, columns=['review'])
clean_data.head()
```
